```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()
```

```python
plt.plot(x, np.sin(x), '-');
```

```python
plt.plot(x, np.cos(x), '--')
```

```python
a = np.linspace(0,10,101)
x = pd.DataFrame(a,columns=['A'])
x['B'] = np.sin(x['A'])
x['C'] = np.cos(x['A'])
x.head()

for col in x.columns:
    plt.plot(x['A'], x[col], color='blue')
plt.legend()
```

```python
plt.plot(x.index, x.B, 's', color='green')
plt.plot(x.index, x.C, 'd', color='blue');
```

```python
plt.style.use('ggplot')
plt.hist(x.B, bins=50, alpha=.5)
plt.hist(x.C, bins=50, alpha=.5);
```
